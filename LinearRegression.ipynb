# **Training and Testing Data**

To apply Linear Regression to the above data, split the data into training and testing sets. Set a variable X equal to the numerical features of the data and a variable y equal to the "Yearly Amount Spent" column.

Use model_selection.train_test_split from sklearn to split the data into training and testing sets. Set test_size=0.2 and random_state=125


```python
X = data[['Avg Session Length', 'Time on App','Time on Website', 'Length of Membership']]
y = data['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=125)
```

Import Linear Regression and train the model on the training set of data

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
linear.fit(X_train,y_train)
```

```python
print('Coefficients: ', linear.coef_)
```

**Predictions are made using the testing set of data after training the model and plotted**

```python
predictions = linear.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

**We can evaluate the model performance by finding the Mean Squared Error and R^2 score**

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print('Mean Squared Error:', mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, predictions)))
print('R^2 Score: ',r2_score(y_test, predictions))
```
